```python
from sympy import diff,symbols, cos, sin
from sympy.solvers import solve, nsolve, solveset
```

```python
x, y = symbols('x y')
```

```python
f = -12*x**4*sin(cos(x)) -18*x**3 + 5*x**2 +10*x -30
```

```python
from sympy.plotting import plot as symplot
symplot(f)
```

```python
symplot(f, (x,-100,100))
```

```python
symplot(f, (x,-19,-1))
```

```python
# ,   
from sympy.solvers import nsolve
roots = [nsolve(f, x, -18),
 nsolve(f, x, -15),
 nsolve(f, x, -12),
 nsolve(f, x, -8),
 nsolve(f, x, -6),
 nsolve(f, x, -2),
 nsolve(f, x, 2),
 nsolve(f, x, 4),
 nsolve(f, x, 8),
 nsolve(f, x, 10),
 nsolve(f, x, 14)
]
roots
```

```python
# 
intervals_between_roots = list(zip(roots, roots[1:]))
intervals_between_roots
```

```python
#  > 0
positive_values = intervals_between_roots[::2]
positive_values
```

```python
#  < 0
negative_values = intervals_between_roots[1:][::2]
negative_values
```

```python
#
dif = diff(-12*x**4*sin(cos(x)) -18*x**3 + 5*x**2 +10*x -30, x)
symplot(dif)
```

```python
symplot(dif, (x,-100,100))
```

```python
symplot(dif, (x,15,20))
```

```python
#   ( ),   
dif_roots = [
 nsolve(dif, x, -19),
 nsolve(dif, x, -16),
 nsolve(dif, x, -13),
 nsolve(dif, x, -10),
 nsolve(dif, x, -7),
 nsolve(dif, x, -4),
 nsolve(dif, x, -1),
 nsolve(dif, x, 0.5),
 nsolve(dif, x, 1.7),
 nsolve(dif, x, 4),
 nsolve(dif, x, 7),
 nsolve(dif, x, 10),
 nsolve(dif, x, 13),
 nsolve(dif, x, 20)
]
dif_roots
```

```python
#  
intervals_between_dif_roots = list(zip(dif_roots, dif_roots[1:]))
intervals_between_dif_roots
```

```python
#  
dif_positive_values = intervals_between_dif_roots[::2]
dif_positive_values
```

```python
#  
dif_negative_values = intervals_between_dif_roots[1:][::2]
dif_negative_values
```
